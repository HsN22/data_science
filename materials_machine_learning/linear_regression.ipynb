## Fitting a baseline model, linear regression to map the control parameters (all parameters *except* `gllbsc_gap`) to the dependent parameter `gllbsc_gap`

**Performing a multiple linear regression with a test set of 20% and train set of 80%.**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

x = df.loc[:, df.columns != "gllbsc_gap"].values
y = df.loc[:, df.columns == "gllbsc_gap"].values

#using a standard scaler to normalise data before training the model

scaler_x = StandardScaler()
x = scaler_x.fit_transform(x)
scaler_y = StandardScaler()
y = scaler_y.fit_transform(y.reshape(-1, 1))

xtrain, xtest, ytrain, ytest = train_test_split(x, y, random_state=0,
                                  train_size=0.8)
regr = LinearRegression()
regr.fit(xtrain, ytrain)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
ypred = regr.predict(xtest)

print('Mean squared error:', mean_squared_error(ypred, ytest))
print('Root mean squared error:', mean_squared_error(ypred, ytest, squared=False))
print('r-squared:', r2_score(ytest, ypred))
```

**Eneg dev seems to increase the band gap the most due to the positive coefficient whilst HOMO_energy seems to decrease the band gap the most due to the negative coefficient as seen below.**

```python
# Coefficients for each feature
coefficients_dict = {
    'Weight dev': -1.31367003e-02,
    'Eneg dev': 3.95650518e+00,
    'NdValence dev': -7.22628425e-01,
    'MeltT dev': -2.52836143e-04,
    'Periodic nature': -2.04341079e-02,
    'GS mean': -3.66958351e-02,
    'GS dev': 1.08873247e-01,
    'MeltT mean': 8.17029949e-04,
    'NValence mean': 4.26555947e-02,
    'HOMO_energy': -1.37984267e+01,
    'Mendeleev dev': -1.05413805e-02,
    'CovRad dev': -9.15536291e-02
}

# Sort coefficients by values in descending order
sorted_coefficients = sorted(coefficients_dict.items(), key=lambda x: x[1], reverse=True)

# Print coefficients in descending order
print("Coefficients in Descending Order:")
for feature, coefficient in sorted_coefficients:
    print(f"{feature}: {coefficient}")
```

**From looking at the linear regression model, the feature that has the greatest influence on band gap in terms of absolute value is HOMO_energy, as seen below.**

```python
# Sort coefficients by absolute values in descending order by ABSOLUTE VALUE
sorted_coefficients = sorted(coefficients_dict.items(), key=lambda x: abs(x[1]), reverse=True)

# Print coefficients in descending order
print("Coefficients in Descending Order:")
for feature, coefficient in sorted_coefficients:
    print(f"{feature}: {abs(coefficient)}")
```

```python
min_gllbsc_gap = df['gllbsc_gap'].min()
max_gllbsc_gap = df['gllbsc_gap'].max()
rangegap = max_gllbsc_gap - min_gllbsc_gap
print("Minimum gllbsc_gap value:", min_gllbsc_gap)
print("Maximum gllbsc_gap value:", max_gllbsc_gap)
print("Range:", rangegap)
```

**The r-squared value is 0.38, suggesting that the independant variables explain somewhat of the variation in the dependant variable, gllbsc_gap. The residual plot suggests there is correlation. ## Developing a Gradient Boosted Regressor to the Same Data

```python
from sklearn.ensemble import GradientBoostingRegressor

boost = GradientBoostingRegressor()
boost.fit(xtrain, ytrain.ravel())
predictions = boost.predict(xtest)
```

```python
plt.scatter(ytest, predictions)
plt.xlabel('True values')
plt.ylabel('Predictions')
```

**The residual plot suggests that the model shows correlation on the test set, there is good performance and accuracy.**

```python
boost.n_estimators
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print('Mean squared error:', mean_squared_error(predictions, ytest))
print('Root mean squared error:', mean_squared_error(predictions, ytest, squared=False))
print('r-squared:', r2_score(ytest, predictions))
```

**Having calculated the metrics after using a gradient boosted regressor, it can be seen that r-squared value is higher, 0.576, and the mean squared error and root mean squared error are lower. This mean that the outcomes from model has improved a lot when using the gradient boosted regressor and the performance of the model is better.**

## Hyperparameter Tuning

```python
boost.get_params().keys()
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    "n_estimators": [50, 100, 200, 500, 1000],
}

search_cv = GridSearchCV(
    GradientBoostingRegressor(learning_rate=0.1), param_grid=param_grid,
    scoring="neg_mean_absolute_error",  n_jobs=2, cv=10
)

search_cv.fit(xtrain, ytrain.ravel())

print(search_cv.best_params_)
```

```python
param_grid = {
    "max_depth":range(1, 10, 2),
    "min_samples_split":range(2, 50, 10)
}

search_cv = GridSearchCV(
    GradientBoostingRegressor(learning_rate=0.1, n_estimators=1000), param_grid=param_grid,
    scoring="neg_mean_absolute_error",  n_jobs=2, cv=10
)

search_cv.fit(xtrain, ytrain.ravel())

print(search_cv.best_params_)
```

```python
param_grid = {
    "learning_rate":[0.01, 0.05, 0.1, 0.2]
}

search_cv = GridSearchCV(
    GradientBoostingRegressor(n_estimators=1000, max_depth=5, min_samples_split=12), param_grid=param_grid,
    scoring="neg_mean_absolute_error",  n_jobs=2, cv=10
)

search_cv.fit(xtrain, ytrain.ravel())

print(search_cv.best_params_)
```

```python
search_cv.best_params_
```

```python
best_regr = search_cv.best_estimator_

predictions = best_regr.predict(xtest)
plt.scatter(ytest, predictions)
plt.xlabel('True values')
plt.ylabel('Predictions')
```

**The residual plot for the model with tuning also shows correlation.**

```python
print('Mean squared error:', mean_squared_error(predictions, ytest))
print('Root mean squared error:', mean_squared_error(predictions, ytest, squared=False))
print('r-squared:', r2_score(ytest, predictions))
```

**After hyperparameter tuning, the mean squared error and the root mean squared error has slightly decreased. The r-squared value has increased. These metrics tell us that performance of the new model has increased with hyperparameter tuning compared to the baseline model and the gradient boosted regressor model.**
